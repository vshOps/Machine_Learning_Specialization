```python
def compute_cost(x,y,w,b):
    #number of training examples
    m = x.shape[0]
    
    cost_sum = 0
    for i in range(m):
        f_wb = w*x[i] + b
        cost = (f_wb - y[i]) ** 2
        cost_sum = cost_sum + cost
    total_cost = (1/(2*m)) * cost_sum
    return total_cost
```

```python
# training data
x = np.array([1,2])
y = np.array([300,500])
# parameters
w = 100
b = 100
cost = compute_cost(x,y,w,b)
print(cost)

```
