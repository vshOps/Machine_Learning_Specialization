```python
#x_train is the iput variable (size in 1000 sqare feet)
#y_train is the target (prize in 1000s of dollars)
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
# plot the datat points
plt.scatter(x_train, y_train, marker = 'x', color ='brown')
#set the title
plt.title("Housing prices")
#set the y_axis lable
plt.xlabel('Size (1000 sqft)')
plt.show()
```

```python
w = 100
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

```python
def compute_model_output(x, w, b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range (m):
        f_wb[i] = w * x[i] + b
        
    return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train, w, b,)
#plot our model prediction 
plt.plot(x_train, tmp_f_wb, color= 'blue', label = ' Our Prediction')
#plot the data points
plt.plot(x_train, y_train, color = 'orange', label='Actual values')
# Pink crosses on actual points
plt.scatter(x_train, y_train, marker='x', color='purple')
# set the title
plt.title("Housing prices")
#Set the y-axis label
plt.ylabel('Price (in 1000s of  Dollars)')
#Set the x-axis label
plt.xlabel('size (1000 sqft)')
plt.legend()
plt.show()
```
